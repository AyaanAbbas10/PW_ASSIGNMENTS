```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
#ANS
#The LookupError exception is used because it forms the base class for all exceptions
#that are raised when an index or a key is not found for a sequence or dictionary 
#You can use LookupError exception class to handle both IndexError and KeyError exception classes.
#KeyError example
try:
    d={"key1":"value1","key3":"value3"}
    print(d["key2"])
except KeyError as e:
    print(e)
#IndexError Example
try:
    l=[1,2,3]
    print(l[5])
except IndexError as e:
    print(e)
```
