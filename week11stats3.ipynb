# Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
# ANS Estimation in statistics refers to the process of using sample data to make inferences or predictions about a population parameter. Estimation involves providing a numerical value that serves as an estimate of the true, unknown population parameter. There are two main types of estimation: point estimation and interval estimation.

# Point Estimate:A point estimate is a single value that is calculated from the sample data and used to approximate the unknown population parameter.
# Interval Estimate (Confidence Interval):An interval estimate is a range of values that is likely to contain the true population parameter with a certain level of confidence.

```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    margin_of_error = 1.96 * (sample_std_dev / (sample_size ** 0.5))  
    lower_limit = sample_mean - margin_of_error
    upper_limit = sample_mean + margin_of_error
    return lower_limit, upper_limit

sample_mean = 50.0
sample_std_dev = 10.0
sample_size = 100

lower_limit, upper_limit = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print(f"Estimated population mean: {sample_mean}")
print(f"Confidence Interval: [{lower_limit}, {upper_limit}] at 95% confidence level")
```

# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

# ANS Hypothesis testing is a fundamental concept in statistics that involves making decisions about population parameters based on sample data. It's a systematic process used to evaluate whether a certain hypothesis about a population parameter is supported by the evidence provided by the sample.

# Null Hypothesis (H0): This is the default or initial assumption. It states that there is no significant effect, relationship, or difference between groups or variables.

# Alternative Hypothesis (H1): This is the hypothesis we want to test. It states that there is a significant effect, relationship, or difference between groups or variables

# Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
# ANS NULL HYPOTHESIS-The average weight of male college student is greater than the average weight of female college students 
# ALTERNATIVE HYPOTHESIS-The average weight of male college students is not greater than the average weight of female college students

```python
# Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
# given a sample from each population.
import scipy.stats as stats

def independent_samples_t_test(sample1, sample2, alpha=0.05):
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    
    if p_value < alpha:
        hypothesis_result = "Reject null hypothesis"
    else:
        hypothesis_result = "Accept null hypothesis"
    return t_statistic, p_value, hypothesis_result

sample1 = [27, 28, 30, 31, 29]
sample2 = [33, 31, 32, 34, 35]
alpha = 0.05

t_statistic, p_value, result = independent_samples_t_test(sample1, sample2, alpha)
print("Results of Hypothesis Test:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
print(f"Conclusion: {result}")
```

 2) Students who receive tutoring have significantly higher average test scores than those who do not.

# Q8. Define p-value and explain its significance in hypothesis testing.
# ANS The p-value is a crucial concept in hypothesis testing. It's a statistical measure that quantifies the strength of the evidence against the null hypothesis. The p-value provides a way to determine whether the observed sample data is consistent with the null hypothesis or whether it suggests that the null hypothesis should be rejected in favor of the alternative hypothesis.
# SIGNIFICANCE OF P-VALUE IN Hypothesis testing
1) Calculation of the p-value:The p-value is calculated based on the observed sample data and the assumed null hypothesis.
- In simpler terms, it indicates how likely the observed data is if the null hypothesis is correct.

2) Interpreting the p-value:A small p-value (typically less than the chosen significance level, α) suggests that the observed data is unlikely to have occurred by random chance alone under the null hypothesis.
- A small p-value indicates strong evidence against the null hypothesis and suggests that you should consider rejecting it.

3) Significance Level (α):The significance level, often denoted by α, is the threshold below which you decide to reject the null hypothesis.
- Common significance levels are 0.05 (5%) or 0.01 (1%).
- If the calculated p-value is less than α, you have evidence to reject the null hypothesis.

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
# random samples of equal size and a null hypothesis that the population means are equal.
import scipy.stats as stats
sample1 = [10.5, 12.8, 13.2, 11.7, 14.3]
sample2 = [9.7, 11.1, 11.8, 10.3, 12.6]

null_hypothesis = "Population means are equal"

t_statistic, p_value = stats.ttest_ind(sample1, sample2)

alpha = 0.05

if p_value < alpha:
    hypothesis_result = "Reject null hypothesis"
else:
    hypothesis_result = "Accept null hypothesis"


print("Two-Sample t-Test for Independent Samples")
print("Sample 1:", sample1)
print("Sample 2:", sample2)
print("Null Hypothesis:", null_hypothesis)
print("Calculated t-statistic:", t_statistic)
print("Calculated p-value:", p_value)
print("Significance level (alpha):", alpha)
print("Conclusion:", hypothesis_result)
```

# Q12: What is t-statistic? State the formula for t-statistic.
# ANS The t-statistic, also known as the Student's t-statistic, is a measure used in hypothesis testing to determine whether there is a significant difference between the sample mean and a hypothesized population mean. It quantifies the difference between the sample mean and the population mean in terms of the standard error of the sample mean.

# The formula for calculating the t-statistic is as follows:
# t=(Sample Mean−Hypothesized Population Mean)/Standard Error of the Sample Mean

 

# Where:Sample Mean is the average of the sample data.
Hypothesized Population Mean is the value assumed in the null hypothesis.

Standard Error of the Sample Mean is the standard deviation of the sample divided by the square root of the sample size (n).

```python
# Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
# sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
# Estimate the population mean revenue with a 95% confidence interval.
import scipy.stats as stats

sample_mean = 500  
sample_std_dev = 50  
sample_size = 50  

standard_error = sample_std_dev / (sample_size ** 0.5)

confidence_level = 0.95
degrees_of_freedom = sample_size - 1
critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_of_freedom)

margin_of_error = critical_value * standard_error

lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

print(f"Estimated population mean revenue: ${sample_mean}")
print(f"95% Confidence Interval: ${lower_limit:.2f} to ${upper_limit:.2f}")
```

```python
# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
# clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
# standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
import scipy.stats as stats

sample_mean = 8
hypothesized_mean = 10
sample_std_dev = 3
sample_size = 100
alpha = 0.05

t_statistic = (sample_mean - hypothesized_mean) / (sample_std_dev / (sample_size ** 0.5))
degrees_of_freedom = sample_size - 1

critical_value = stats.t.ppf(alpha, degrees_of_freedom)

if t_statistic < critical_value:
    hypothesis_result = "Reject null hypothesis"
else:
    hypothesis_result = "Accept null hypothesis"

print("Results of Hypothesis Test:")
print("Hypothesized Mean:", hypothesized_mean)
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
print("Sample Size:", sample_size)
print("Degrees of Freedom:", degrees_of_freedom)
print("Calculated t-statistic:", t_statistic)
print("Critical Value:", critical_value)
print("Conclusion:", hypothesis_result)
```

```python
# Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
# 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
# score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
# groups are equal with a significance level of 0.01.
import scipy.stats as stats

sample_mean_1 = 80
sample_std_dev_1 = 10
sample_size_1 = 30

sample_mean_2 = 75
sample_std_dev_2 = 8
sample_size_2 = 40

alpha = 0.01


pooled_std_dev = ((sample_std_dev_1 ** 2 / sample_size_1) + (sample_std_dev_2 ** 2 / sample_size_2)) ** 0.5


t_statistic = (sample_mean_1 - sample_mean_2) / pooled_std_dev

degrees_of_freedom = sample_size_1 + sample_size_2 - 2

critical_value = stats.t.ppf(1 - (alpha / 2), degrees_of_freedom)

if abs(t_statistic) > critical_value:
    hypothesis_result = "Reject null hypothesis"
else:
    hypothesis_result = "Accept null hypothesis"

print("Results of Hypothesis Test:")
print("Sample Mean (Group 1):", sample_mean_1)
print("Sample Mean (Group 2):", sample_mean_2)
print("Sample Standard Deviation (Group 1):", sample_std_dev_1)
print("Sample Standard Deviation (Group 2):", sample_std_dev_2)
print("Sample Size (Group 1):", sample_size_1)
print("Sample Size (Group 2):", sample_size_2)
print("Degrees of Freedom:", degrees_of_freedom)
print("Calculated t-statistic:", t_statistic)
print("Critical Value:", critical_value)
print("Conclusion:", hypothesis_result)
```
