```python
#Q1. How can you create a Bokeh plot using Python code?
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 8]

source = ColumnDataSource(data={'x': x, 'y': y})
plot = figure(title="Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
plot.circle('x', 'y', source=source, size=10, color='blue', legend_label='Data Points')
output_notebook() 
show(plot)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
# Glyphs are visual shapes or markers that represent data points on a plot. 
# Glyphs are the building blocks of visualizations, such as scatter plots, line charts, bar charts.
# Each glyph represents a data point or a collection of data points and you can customize their 
# properties to control how they appear on the plot.
# the plot.circle() function is used to add a circle glyph to the plot

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
plot = figure(title="Customized Plot Example", x_axis_label='X-axis', y_axis_label='Y-axis')
plot.circle(x, y, size=10, color='blue', legend_label='Data Points')

# Customize plot appearance
plot.title.text_color = "navy"
plot.title.text_font_size = "20px"
plot.title.align = "center"

# Customize axis appearance
plot.xaxis.axis_label_text_font_size = "14pt"
plot.xaxis.axis_label_text_color = "purple"
plot.yaxis.axis_label_text_font_size = "14pt"
plot.yaxis.axis_label_text_color = "purple"

# Customize legend appearance
plot.legend.title = "Legend Title"
plot.legend.title_text_font_size = "16pt"
plot.legend.label_text_font_size = "12pt"


output_notebook()  
show(plot)
```
